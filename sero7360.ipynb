    for i from 2 to n-1:

`````python
def factorize(n):
    highest_factor = n//2               # Only have to go as high as half of the number we are factoring 
    if n <=1:
        return False
    for i in range(2,highest_factor):
        if n % i == 0 :
            return i, n // i            # return a tuple of p ,q 
    return False
`````
